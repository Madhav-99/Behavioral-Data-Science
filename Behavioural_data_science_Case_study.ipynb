```python
df=pd.read_excel('Copy of Behavioural Data Science Assignment.xlsx',sheet_name='Sheet9')
df.drop(columns=['Unnamed: 1','Unnamed: 2'],inplace=True)
df.head()
```

```python
#Renaming the column entries to better understand the data
param=['Efforts given','importance_effort','effort_learning_development_of_core_tech_skills','importance_effort_learning_development_of_core_tech_skills'
        ,'effort_in_other_domain','importance_effort_in_other_domain','effort_in_communication_Emotional_intelligence','importance_effort_in_communication_Emotional_intelligence'
        ,'effort_in_efficiency','importance_effort_in_efficiency']
```

```python
df['Parameters']=param
```

```python
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = df['Parameters']
values1 = df['Neha']
values2 = df['Amit']
values3 = df['Sachin']

# Set the width of each bar
bar_width = 0.25

# Set the index for the x-axis
index = np.arange(len(categories))

# Set the figure size
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting the bars
rects1 = ax.bar(index - bar_width, values1, bar_width, label='Neha')
rects2 = ax.bar(index, values2, bar_width, label='Amit')
rects3 = ax.bar(index + bar_width, values3, bar_width, label='Sachin')

# Set x-axis labels
ax.set_xticks(index)
ax.set_xticklabels(categories)

# Set labels and title
ax.set_xlabel('Parameters')
ax.set_ylabel('Points')
ax.set_title('Side-by-Side Bar Graph of all 3 candidates')

# Add legend
ax.legend()

# Show the graph
plt.xticks(rotation=90)
plt.show()
```

1. Here we can see a maximum amount of effort seen from Neha in comparison to Amit and Sachin
2. Amit is consistently in between Neha and Sachin

```python
#Sachin
num_bars=10
colors = ['blue', 'orange'] * (num_bars // 2)
fig, ax = plt.subplots(figsize=(12, 6))
plt.barh(df['Parameters'],df['Sachin'],color=colors)
legend_labels = ['Effort', 'importance of the effort']  
legend_handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in colors[:2]]
plt.legend(legend_handles, legend_labels)
plt.show()
```

 1.   **From the above bar graph we can see that it is important for sachin to learn the core technical skills and giving effort in his job but the effort that he is putting is not enough**

2.   **In contrast to that he is putting effort for communication skills and emotional intelligence but it is not important to him.**


*   Maybe the reason Sachin is happy is due to his Emotional Intelligence
*   He might be happy due to the fact that he has found his interest in other domain(effort in other domain and importance in other domains are similar)
* He considers himself efficient too.
